```python
print(df.isnull().sum())
```

```python
df.fillna(0, inplace=True)
```

```python
df['Cost ($)'] = pd.to_numeric(df['Cost ($)'])
```

```python
df['Week'] = pd.to_datetime(df['Week'])
```

```python
df
```

```python
df['Payment ($)'] = pd.to_numeric(df['Payment ($)'].str.replace('[\$,]', '', regex=True))
```

```python
df
```

```python
df['Payment Date'] = pd.to_datetime(df['Payment Date'])
```

```python
df
```

```python
df['Payment ($)'].fillna(0, inplace=True)
```

```python
df
```

```python
df['ROAS'] = df['Payment ($)'] / df['Cost ($)']
```

```python
df
```

```python
df['CTR'] = (df['Clicks'] / df['Impressions']) * 100
```

```python
df
```

```python
df['Conversion_Rate'] = (df['Prospects'] / df['Clicks']) * 100
```

```python
df
```

```python
# Group by week and month, and calculate mean for the key metrics
weekly_report = df.resample('W', on='Week').mean()
monthly_report = df.resample('M', on='Week').mean()
```

```python
weekly_report = weekly_report[['Cost ($)', 'ROAS', 'CTR', 'Conversion_Rate']]
monthly_report = monthly_report[['Cost ($)', 'ROAS', 'CTR', 'Conversion_Rate']]
```

```python
weekly_report
```

```python
monthly_report
```

```python
monthly_report['Month'] = monthly_report.index.strftime('%B %Y')
monthly_report = monthly_report[['Month', 'Cost ($)', 'ROAS', 'CTR', 'Conversion_Rate']]
monthly_report
```

```python
weekly_report['Week of Year'] = weekly_report.index.strftime('%U, %Y')
weekly_report = weekly_report[['Week of Year', 'Cost ($)', 'ROAS', 'CTR', 'Conversion_Rate']]
weekly_report
```

```python
#Identify the Most Profitable Channel
#Comparing the ROAS for Google Ads and the listing site campaigns.
most_profitable_channel = df.groupby('Campaign')['ROAS'].mean().idxmax()
most_profitable_channel
```

```python
#Identify the Most Profitable Category/Keyword
#Analyze the data to find the category or keyword with the highest ROAS.
most_profitable_category = df.groupby('Ad group')['ROAS'].mean().idxmax()
most_profitable_category
```

```python
# Set the plot style
sns.set_style('whitegrid')

# Creating time series graphs for the metrics
plt.figure(figsize=(12, 6))
sns.lineplot(x='Week', y='ROAS', data=df)
plt.title('ROAS over Time')
plt.xlabel('Week')
plt.ylabel('ROAS')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Week', y='CTR', data=df)
plt.title('CTR over Time')
plt.xlabel('Week')
plt.ylabel('CTR')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Week', y='Conversion_Rate', data=df)
plt.title('Conversion Rate over Time')
plt.xlabel('Week')
plt.ylabel('Conversion Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
#Seasonality Analysis: # Seasonal decomposition of ROAS using moving averages
from statsmodels.tsa.seasonal import seasonal_decompose
df.dropna(subset=['ROAS'], inplace=True)

result = seasonal_decompose(df['ROAS'], model='additive', period=7)
result.plot()
plt.show()
```

```python
#Trend Analysis:
df['CTR_rolling_mean'] = df['CTR'].rolling(window=5, min_periods=1).mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Week', y='CTR', data=df, label='CTR')
sns.lineplot(x='Week', y='CTR_rolling_mean', data=df, label='5-day Rolling Avg CTR')
plt.title('CTR over Time with 5-day Rolling Average')
plt.xlabel('Week')
plt.ylabel('CTR')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#Anomaly Detection:
from scipy import stats
df['Conversion_Rate_Zscore'] = stats.zscore(df['Conversion_Rate'])
anomalies = df[df['Conversion_Rate_Zscore'].abs() > 3]
plt.figure(figsize=(12, 6))
sns.lineplot(x='Week', y='Conversion_Rate', data=df, label='Conversion Rate')
sns.scatterplot(x='Week', y='Conversion_Rate', data=anomalies, color='red', label='Anomalies')
plt.title('Conversion Rate over Time with Anomalies')
plt.xlabel('Week')
plt.ylabel('Conversion Rate')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#Correlation Analysis:
correlation_matrix = df[['ROAS', 'CTR', 'Conversion_Rate']].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Insights from keyword type:
# Plot the metrics for each keyword type
# Group by 'Keyword Type' and calculate the mean for key metrics
keyword_type_metrics = df.groupby('Keyword type').mean()[['ROAS', 'CTR', 'Conversion_Rate']]

plt.figure(figsize=(10, 6))
sns.barplot(x=keyword_type_metrics.index, y='ROAS', data=keyword_type_metrics)
plt.title('ROAS by Keyword Type')
plt.xlabel('Keyword Type')
plt.ylabel('ROAS')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=keyword_type_metrics.index, y='CTR', data=keyword_type_metrics)
plt.title('CTR by Keyword Type')
plt.xlabel('Keyword Type')
plt.ylabel('CTR')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=keyword_type_metrics.index, y='Conversion_Rate', data=keyword_type_metrics)
plt.title('Conversion Rate by Keyword Type')
plt.xlabel('Keyword Type')
plt.ylabel('Conversion Rate')
plt.show()
```

```python
#The most search keyword
search_keyword_counts = df['Search Keyword'].value_counts()
most_searched_keyword = search_keyword_counts.index[0]
most_searched_keyword_count = search_keyword_counts.iloc[0]
print("The most searched keyword is:", most_searched_keyword)
print("Number of times it was searched:", most_searched_keyword_count)
```

```python
#Visualising the Search Keywords
search_keyword_counts = df['Search Keyword'].value_counts()
top_n = 10  # top 10 keywords
top_search_keywords = search_keyword_counts.head(top_n)
other_search_keywords_count = search_keyword_counts.sum() - top_search_keywords.sum()
top_search_keywords['Others'] = other_search_keywords_count

plt.figure(figsize=(10, 10))
plt.pie(top_search_keywords, labels=top_search_keywords.index, autopct='%1.1f%%', startangle=140)
plt.title('Most Searched Keywords')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is drawn as a circle.
plt.show()
```

```python
#Which ad group gained the most impressions
category_impressions = df.groupby('Ad group')['Impressions'].sum()
category_impressions = category_impressions.sort_values(ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(category_impressions.index, category_impressions.values)
plt.xlabel('Ad group')
plt.ylabel('Impressions')
plt.title('Impressions by Ad Category')
plt.xticks(rotation=45, ha='right')
plt.show()
```
